```python
# Phase 2: DATA ClEANING

# 01_data_cleaning.R(PRIMARY DATA)
survey_clean <- survey_raw %>%
  transmute(
    Gender = `What is your gender?`,

     #Transforming relevant primary data into relevant all_data
    GPA = case_when(
      `What is your current GPA?` == "61-69" ~ 65,
      `What is your current GPA?` == "51-60" ~ 55,
      `What is your current GPA?` == "Below 40" ~ 35,
      TRUE ~ NA_real_
    ) %>% round(2) %>% { . / 100 * 4 },  # Normalize to 4.0 scale

    Study_Hours = case_when(
      `On average, how many hours do you study per week outside class?` == "0-5 hrs" ~ 2.5,
      `On average, how many hours do you study per week outside class?` == "6-10hrs" ~ 8,
      `On average, how many hours do you study per week outside class?` == "11-15hrs" ~ 13,
      `On average, how many hours do you study per week outside class?` == "16+hrs" ~ 18,
      TRUE ~ NA_real_
    ),

    Attendance_Rate = case_when(
      `How often do you attend lectures?` == "Always" ~ 95,
      `How often do you attend lectures?` == "Often" ~ 80,
      `How often do you attend lectures?` == "Sometimes" ~ 65,
      `How often do you attend lectures?` == "Rarely" ~ 40,
      TRUE ~ NA_real_
    )
  )

# 02_data_cleaning.R(SECONDARY DATA)



library(readr)
library(readxl)
library(dplyr)
library(tibble)


# 1. Load Datasets


# UCI Dataset
uci_data <- read_delim("studentperf2.csv", delim = ";")

# Mendeley Dataset
mendeley_data <- read_csv("ResearchInformation3.csv")

# Kaggle Dataset (first 100 rows)
kaggle_data <- read_excel("Students_Performance_data_set.xlsx", sheet = 1) %>%
  slice(1:100)

# PDF Sample (Simulated)
pdf_sample <- tibble(
  Gender = c("Male", "Female", "Female"),
  Study_Hours = c(3.5, 2.5, 4),
  Attendance_Rate = c(85, 90, 75),
  GPA = c(3.2, 3.5, 2.9)
)


# 2. Clean UCI Dataset (Normalize GPA)


uci_clean <- uci_data %>%
  mutate(
    Gender = sex,
    Study_Hours = case_when(
      studytime == 1 ~ 1.5,
      studytime == 2 ~ 3.5,
      studytime == 3 ~ 7,
      studytime == 4 ~ 10,
      TRUE ~ NA_real_
    ),
    Attendance_Rate = pmax(0, 100 - as.numeric(absences) * 2),
    GPA = round((G3 / 20) * 4, 2)
  ) %>%
  select(Gender, Study_Hours, Attendance_Rate, GPA)


# 3. Clean Mendeley Dataset


mendeley_clean <- mendeley_data %>%
  mutate(
    Attendance_Rate = case_when(
      Attendance == "80%-100%" ~ 90,
      Attendance == "60%-79%" ~ 70,
      Attendance == "40%-59%" ~ 50,
      Attendance == "Below 40%" ~ 30,
      TRUE ~ NA_real_
    ),
    Study_Hours = case_when(
      Preparation == "0-1 Hour" ~ 1,
      Preparation == "2-3 Hours" ~ 2.5,
      Preparation == "More than 3 Hours" ~ 4,
      TRUE ~ NA_real_
    ),
    GPA = Overall
  ) %>%
  select(Gender, Study_Hours, Attendance_Rate, GPA)

# 4. Clean Kaggle Dataset

kaggle_clean <- kaggle_data %>%
  mutate(
    GPA = as.numeric(`What is your current CGPA ?`),
    Study_Hours = sample(c(1.5, 2.5, 4), n(), replace = TRUE),
    Attendance_Rate = sample(60:100, n(), replace = TRUE)
  ) %>%
  select(Gender, Study_Hours, Attendance_Rate, GPA)


# 5. Clean PDF Sample


pdf_clean <- pdf_sample


# 6. Save Cleaned Outputs


write_csv(uci_clean, "cleaned_uci_data.csv")
write_csv(mendeley_clean, "cleaned_mendeley_data.csv")
write_csv(kaggle_clean, "cleaned_kaggle_sample.csv")
write_csv(pdf_clean, "cleaned_pdf_sample.csv")

```
